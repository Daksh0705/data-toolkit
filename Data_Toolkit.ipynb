```python
# How do you create a 2D NumPy array and calculate the sum of each row?
import numpy as np

# Step 1: Create a 2D NumPy array
array_2d = np.array([
    [10, 20, 30],
    [5, 15, 25],
    [1, 2, 3]
])

# Step 2: Calculate the sum of each row using axis=1
row_sums = np.sum(array_2d, axis=1)

# Step 3: Print the original array and the row sums
print("2D Array:")
print(array_2d)

print("\nSum of each row:")
print(row_sums)
```

```python
#Write a Pandas script to find the mean of a specific column in a DataFrame.
import pandas as pd

# Step 1: Create a sample DataFrame
data = {
    'Name': ['daksh', 'anuj', 'pawan', 'navneet'],
    'Age': [25, 30, 35, 40],
    'Salary': [50000, 60000, 55000, 65000]
}

df = pd.DataFrame(data)

# Step 2: Calculate the mean of the 'Salary' column
mean_salary = df['Salary'].mean()

# Step 3: Print the DataFrame and the mean salary
print("DataFrame:")
print(df)

print("\nMean Salary:")
print(mean_salary)
```

```python
#Create a scatter plot using Matplotlib.
import matplotlib.pyplot as plt

# Step 1: Define the data points
x = [10, 20, 30, 40, 50]
y = [15, 25, 35, 30, 45]

# Step 2: Create the scatter plot
plt.scatter(x, y, color='red', marker='o')

# Step 3: Add labels and title
plt.xlabel('X Axis - Example Values')
plt.ylabel('Y Axis - Measured Values')
plt.title('Simple Scatter Plot')

# Step 4: Show the plot
plt.grid(True)
plt.show()
```

```python
# How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Create a sample DataFrame
data = {
    'Math': [90, 85, 78, 92, 88],
    'Science': [85, 80, 75, 90, 86],
    'English': [70, 75, 80, 65, 72]
}
df = pd.DataFrame(data)

# Step 2: Calculate the correlation matrix
correlation_matrix = df.corr()

# Step 3: Visualize using a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
#Write a program to perform element-wise multiplication of two NumPy arrays.
import numpy as np

# Step 1: Create two NumPy arrays of the same shape
array1 = np.array([2, 4, 6, 8])
array2 = np.array([1, 3, 5, 7])

# Step 2: Perform element-wise multiplication
result = array1 * array2

# Step 3: Print the input arrays and the result
print("Array 1:", array1)
print("Array 2:", array2)
print("Element-wise Multiplication:", result)
```

```python
#Create a line plot with multiple lines using Matplotlib.
import matplotlib.pyplot as plt

# Step 1: Define the data for multiple lines
x = [0, 1, 2, 3, 4, 5]
y1 = [0, 1, 4, 9, 16, 25]  # y = x^2
y2 = [0, 1, 2, 3, 4, 5]    # y = x
y3 = [0, 1, 1, 2, 3, 5]    # Fibonacci sequence

# Step 2: Create the line plot with multiple lines
plt.plot(x, y1, label='y = x^2', color='blue', marker='o')
plt.plot(x, y2, label='y = x', color='green', marker='x')
plt.plot(x, y3, label='Fibonacci', color='red', marker='s')

# Step 3: Add labels and title
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Multiple Line Plot')

# Step 4: Display the legend
plt.legend()

# Step 5: Show the plot
plt.grid(True)
plt.show()
```

```python
#Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.
import pandas as pd

# Step 1: Create a sample DataFrame
data = {
    'Name': ['daksh', 'anuj', 'satyam', 'lakshya', 'shashwat'],
    'Age': [25, 30, 35, 40, 22],
    'Salary': [50000, 60000, 70000, 80000, 45000]
}
df = pd.DataFrame(data)

# Step 2: Filter rows where the 'Age' is greater than 30
threshold = 30
filtered_df = df[df['Age'] > threshold]

# Step 3: Print the filtered DataFrame
print(filtered_df)
```

```python
#Create a histogram using Seaborn to visualize a distribution.
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Create a sample data (e.g., random ages)
data = [22, 23, 25, 26, 27, 28, 30, 31, 32, 34, 35, 36, 38, 39, 40, 41, 42, 43, 45, 46]

# Step 2: Create the histogram using Seaborn
sns.histplot(data, kde=True, bins=10, color='blue')

# Step 3: Add labels and title
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age Distribution')

# Step 4: Show the plot
plt.show()
```

```python
#Perform matrix multiplication using NumPy.
import numpy as np

# Step 1: Define two matrices
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# Step 2: Perform matrix multiplication
C = np.dot(A, B)  # or use C = A @ B

# Step 3: Print the result
print("Matrix A:")
print(A)
print("\nMatrix B:")
print(B)
print("\nMatrix Multiplication Result (A * B):")
print(C)
```
